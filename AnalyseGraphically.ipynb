```python
#import numpy as np
import pandas as pd
from bqplot import *
```

```python
inputfile = 'TBone-1-inch.txt'
```

```python
all = pd.read_csv (inputfile)
```

```python
all
```

```python
x_sc = LinearScale()
y_sc = LinearScale()
```

```python
ax_x = Axis(label='Elapsed Time', scale=x_sc, grid_lines='solid')
ax_y = Axis(label='Temperature',  scale=y_sc, orientation='vertical', grid_lines='solid')
```

```python
line = Lines(x=all['ElapsedTime'], y=all[''], scales={'x': x_sc, 'y': y_sc})
fig = Figure(axes=[ax_x, ax_y], marks=[line], title='First Example')
fig
```
